```python
weather = pd.get_dummies(weather, columns=['풍향(16방위)', '운형(운형약어)'])
weather.columns
```

```python
weather['풍향(16방위)_0.0'] = weather['풍향(16방위)_0.0'] + weather['풍향(16방위)_360.0']
weather.drop(columns=['풍향(16방위)_360.0'], inplace=True)
```

```python
ways = [0, 20, 50, 70, 90, 110, 140, 160,
        180, 200, 230, 250, 270, 290, 320, 340]
```

```python
for n in ways:
    weather[f'풍향(16방위)_{n}.0'] = weather[f'풍향(16방위)_{n}.0'] * weather['풍속(m/s)']
```

```python
for n in ways:
    weather.rename(columns={f'풍향(16방위)_{n}.0':f'풍속(16방위)_{n}'}, inplace=True)
```
