```python
from imblearn.under_sampling import RandomUnderSampler
cols = df.columns[6:]
X = df[cols]
y = df['next_mask_need']
```

```python
X_sample, y_sample = RandomUnderSampler(random_state=2).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample, columns=cols)
y_samp = pd.DataFrame(data=y_sample, columns=['next_mask_need'])
df_samp = pd.concat([X_samp, y_samp], axis=1)
df_samp['next_mask_need'].value_counts()
```

```python
sns.countplot('next_mask_need', data=df_samp)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
model = DecisionTreeClassifier(random_state=2)
model.fit(X_train, y_train)
print("학습용 :",model.score(X_train, y_train))
print("검증용 :",model.score(X_test, y_test))
```

```python
model = DecisionTreeClassifier(max_depth=3, random_state=2)
model.fit(X_train, y_train)
print("학습용 :",model.score(X_train, y_train))
print("검증용 :",model.score(X_test, y_test))
```

```python
for i in range(1, 11):
    model = DecisionTreeClassifier(max_depth=i, random_state=2)
    model.fit(X_train, y_train)
    print(f"학습용{i} :",model.score(X_train, y_train))
    print(f"검증용{i} :",model.score(X_test, y_test))
```

```python
score = []
for i in range(1, 21):
    model = DecisionTreeClassifier(max_depth=i, random_state=2)
    model.fit(X_train, y_train)
    score.append(model.score(X_test, y_test))
print(max(score))
print(np.argmax(score) + 1)
```
