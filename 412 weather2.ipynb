```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = "C:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family = font_name)
```

```python
df = pd.read_csv("C:/workspace/data/기상정보/weather_preprocessed.csv")
```

```python
df
```

```python
df['일시'] = pd.to_datetime(df['일시'])
df['날짜'] = df['일시'].dt.date
```

```python
df
```

```python
df.columns
```

```python
df_daily = df.groupby('날짜').mean()
df_daily['강수량(mm)'] = df_daily['강수량(mm)'] * 24
df_daily['적설(cm)'] = df_daily['적설(cm)'] * 24
```
